```python
import pickle
def read():
    with open('essen.pkl', 'rb') as f:
        while True:
            try:
                yield pickle.load(f)
            except EOFError:
                break
```

```python
from itertools import islice

data = islice(read(), 5)
```

```python
# sanity check - make sure all subseqs in a pattern match
# make sure patterns metrics make sense
from tqdm import tqdm
data = read()
spansPerPattern = []
spanLengths = []
patternsPerSequence = []
sequenceLengths = []
total = 0
for i, (seq, ann, index) in enumerate(data):
    total += 1
    hashableSeq = tuple((n.midi, n.duration) for n in seq)
    sequenceLengths.append(len(hashableSeq))
    patternsPerSequence.append(len(ann))
    for pattern in ann:
        spansPerPattern.append(len(pattern))
        canonicalSubSeq = hashableSeq[pattern[0][0]:pattern[0][1]]
        for spanStart, spanEnd in pattern[1:]:
            spanLengths.append(spanEnd - spanStart)
            newSubSeq = hashableSeq[spanStart:spanEnd]
            if canonicalSubSeq != newSubSeq: 
                print(i)
                for a, b in zip(canonicalSubSeq, newSubSeq):
                    print(a, b)
                print(ann)
                for i, n in enumerate(hashableSeq):
                    print(i, n)
                print(index)
                raise
            else:
                print(len(canonicalSubSeq), len(newSubSeq), canonicalSubSeq == newSubSeq)
```
